# What Is API: 
API stands for Aplication Programming Interface. Its like a waiter in a restaurant. 

* You (the client) meke a request ("i'd like chicken tacos")

* The waiter(API) takes you request to the kitcheck(server) 

* The kitchen prepares you food (processes the request)

* The waiter brings back what you ordered (the response). 

In technical terms, an API is a set of rules that allows different software application to communicate with each other.

# Types of APIs: 
1. **Web APIs:** Most common type you'll work with (what we'll focus on)

2. **Library APIs:** How you interact with software libraries (like the requests)

3. **Operating System APIs:**
How programs interact with an OS. 

# How Web APIs Work

Web APIs typically use HTTP (the same protocol web browsers use) to send and recieve data. Here's the basic flow: 

1. **Client sends a request** "Give me information about user #123"

2. **Server processess the request:** Looks up user#123 in its database

3. **Server sends a response:** Returns the user data (usually in JSON format)

# HTTP Basics
HTTP defines how messages are formated and transmitted.

## HTTP Methods(Verbs)

*`GET:` Retrieve data (like loading a webpage)
*`POST:` Send data to create something new (like submitting a form).
* `PUT:` Update existing data.

* `DELETE:` Remove data.

* `PATCH:` Partially update data.

## HTTP Status Codes

These tell you if your request succeded or failed: 

* 200s: Success

* 200 Ok - Request succeeded. 

* 201 Created - Resource was created. 

300s: Redirection 
* 301 Moved Permanently. 

400s: Client errors. 

* 400 Bad Request - You request was malformed. 

* 401 Unauthorized - You need to authenticate. 

403 Forbidden - You're not allowed to do this. 

404 Not Found - Resource doesn't exist. 

500s: Server errors. 
- Internal Server Error - Something went wrong on their end. 

## API Endpoints
An endpoint is a specific URL you interact with to perform operations. For example:

https://api.example.com/users - Might give you a list of users

https://api.example.com/users/123 - Might give you data about user with ID 123

````python
import requests 

def weather ( city_name ,api_key , units ="metric"):
        base_url = "https://api.openweathermap.org/data/2.5/weather" 
        params = {
        "q": city_name,
        "appid": api_key,
        "units": units
        }
        
        try : 
            response = requests.get( base_url , params=params)
            response.raise_for_status
            data = response.json()
            return data

        except response.exceptions.HTTPError as errh:
            print(f"HTTP ERROR: {errh}")
            
            if response.status_code == 401: 
                print(f"check your API key either Invalid or Inactivated!")
            
            elif response.status_code == 404:
                print(f"Unable to find the data for the city {city_name}!")
            return None
        except response.exceptions.ConnectionError as errc: 
            print(f"Connection error {errc}")
        except requests.exceptions.Timeout as errt:
            print(f"Timeout Error: {errt}")
            return None
        except requests.exceptions.RequestException as err:
            print(f"An unexpected error occurred: {err}")
            return None   
                
````


> **`import requests` (the Python library)** vs **FastAPI**.


## ‚úÖ The Real Difference:

| Concept          | `requests` (Python library)             | FastAPI (web framework)                             |
| ---------------- | --------------------------------------- | --------------------------------------------------- |
| **What it does** | Sends HTTP requests (client-side)       | Builds APIs that handle HTTP requests (server-side) |
| **Role**         | Client ‚Äì asks for data                  | Server ‚Äì provides data                              |
| **Example use**  | Connects to an API                      | Creates your own API                                |
| **Who uses it?** | Python programs that need external data | Developers building web APIs                        |

---

## üì§ `requests` ‚Üí Client-side (Sends Requests)

* `requests` is a **Python library** for sending HTTP requests to APIs.
* You use it **when you want to talk to an external server or API**.

### Example:

```python
import requests

response = requests.get("https://api.example.com/data")
print(response.json())
```

This code is like saying:

> "Hey, API server, give me some data!"
> You‚Äôre the **client** here.

---

## üõ†Ô∏è FastAPI ‚Üí Server-side (Handles Requests)

* FastAPI is a **Python web framework** used to **build APIs**.
* You use it to **receive and respond to HTTP requests**.

### Example:

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/hello")
def say_hello():
    return {"message": "Hello from the server!"}
```

This code **creates an API** that someone else (or `requests`) can call.

---

## ü§ù How They Work Together

If you build a FastAPI app, you can call it using `requests`:

```python
# client.py (client side)
import requests

response = requests.get("http://localhost:8000/hello")
print(response.json())
```

This sends a request to your **FastAPI server**, which responds with data.

---

## üß† Summary

| `requests`                | FastAPI                |
| ------------------------- | ---------------------- |
| Acts as **client**        | Acts as **server**     |
| Sends requests            | Handles requests       |
| Used in any Python script | Used to build web APIs |
